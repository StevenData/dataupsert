```
# df = spark.read.load('https://asadatalakeyv8r2e0.blob.core.windows.net/demodata/georef-australia-state-suburb.json', format='json')
df = spark.read.load('wasbs://demodata@asadatalakeyv8r2e0.blob.core.windows.net/georef-australia-state-suburb.json', format='json')
df = spark.read.load('abfss://demodata@asadatalakeyv8r2e0.blob.core.windows.net/georef-australia-state-suburb.json', format='json')
df.printSchema()
```

```
df_nogeo=df.select(	'fields.scc_name','fields.lga_name','fields.ste_name','fields.ste_code','fields.lga_code','fields.scc_code'
	,'fields.scc_type','fields.scc_area_code','fields.year'
	,'datasetid','recordid','record_timestamp')
df_nogeo.printSchema()
```

```
df_nogeo.write.json('abfss://demodata@asadatalakeyv8r2e0.dfs.core.windows.net/australia-state-suburb.json', mode='overwrite')
```

```
df_nogeo.show(30,0)
```

```
from pyspark.sql.functions import *
df_favsuburb=(df_nogeo.withColumn("suburb",regexp_extract("scc_name",r'^(.+?)(\s*\(.*\)\s*)?$',1))
	.groupBy('suburb')
    .agg(count('*').alias('cnt'))
    .orderBy(col('cnt').desc())
    .limit(20))
display(df_favsuburb)
```

```
from pyspark.sql.functions import *
df_favsuburb=(df_nogeo.withColumn("suburb",regexp_extract("scc_name",r'^(.+?)(\s*\(.*\)\s*)?$',1))
	.groupBy('suburb')
    .agg(count('*').alias('cnt'),collect_list('ste_name').alias('states'))
    .orderBy(col('cnt').desc())
    .limit(20))
display(df_favsuburb)
```
